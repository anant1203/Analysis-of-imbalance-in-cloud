```python
container_df2 = pd.read_csv('/home/mr_malviya/Desktop/baksaa/Study/Fall2018/Cloud/Final Project/Data/container_event.csv',header=None)
container_df2.columns = ['timestamp','event_type','instance_id','machine_id','cpu_requested','mem_requested','disk_requested','cpu_id','null']
container_df2 = container_df2.drop(columns='null')
container_df2.tail()
```

```python
kdf['cpu_requested'] = container_df2['cpu_requested']
kdf['mem_requested'] = container_df2['mem_requested']
kdf['disk_requested'] = container_df2['disk_requested']
```

```python
dummy = container_df2[['instance_id','cpu_requested']]
trial = container_df1.merge(dummy, how='left')
trial = trial.merge(container_df2[['instance_id','disk_requested']], how='left')
trial = trial.merge(container_df2[['instance_id','mem_requested']], how='left')

trial.loc[trial['instance_id']==106]
```

```python

trial['target_cpu'] = np.ceil((trial['cpu_used']/100)*trial['cpu_requested'])
trial['target_mem'] = ((trial['mem_used']/100)*trial['mem_requested'])
trial['target_disk'] = ((trial['disk_used']/100)*trial['disk_requested'])
trial.loc[trial['instance_id']==106]
```

```python
trial = trial.dropna()
from sklearn.linear_model import LinearRegression

reg = LinearRegression().fit(trial[['instance_id','cpu_requested','mem_requested','disk_requested']], trial[['target_cpu','target_mem','target_disk']])
```

```python
test = trial[['instance_id','cpu_requested','mem_requested','disk_requested']].loc[trial['instance_id']==550]
predictions=reg.predict(test)
for i in predictions:
    print(np.ceil(i[0]), i[1], i[2])

    
#trial['cpu_requested'] = pd.to_numeric(trial['cpu_requested'], downcast='integer')
#trial
```
